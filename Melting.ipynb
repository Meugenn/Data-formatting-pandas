```python
def melt_signals(file_path, sheet_name, nrows, first_col_iterate = 6):
    #---- Getting the file ----
    sheet = pd.read_excel(file_path, 
                        sheet_name=sheet_name, 
                        header=1,
                        nrows=nrows)
#     sheet1_questions = pd.read_excel(r'.\Noki - Internship (Signals To Review) Project 2.xlsx', 
#                                      sheet_name='Sheet1',
#                                      header=None, nrows=1)

    columns_to_melt = [[] for i in range(7)]
    for i in range(first_col_iterate, len(sheet.columns)):
        columns_to_melt[(i-first_col_iterate)%7].append(sheet.columns[i])
        
    #---- Melting data ----
    melted_sheet = pd.DataFrame()
    column_names = [
        'Answer',
        'Details',
        'Quotes',
        'Evidence',
        'Incorrect',
        'Partial / Whole',
        'Inaccuracy reason'
    ]
    i = 0
    for group in columns_to_melt:
        melted_group = pd.melt(sheet, 
                               id_vars = sheet.columns[list(range(0, first_col_iterate))], 
                               value_vars=group, 
                               value_name=column_names[i])
        if i == 0:
            melted_sheet = melted_group
        else:
            melted_sheet[column_names[i]] = melted_group[column_names[i]]
        i += 1
    melted_sheet.rename(columns={'variable': 'Question'}, inplace=True)
    melted_sheet.head()
    
# #     ---- Processing data ----
    melted_sheet['Incorrect'] = melted_sheet['Incorrect'].apply(lambda x: False if x in ['False ', 'FALSE ', 0] else
                                                              (True if x == 1 else
                                                              ("" if np.isnan(x) else x)))
    
    
    return melted_sheet
```

```python
path = r'.\Noki - Internship (Signals To Review) Project 2.xlsx'
sheet1 = melt_signals(path, 'Sheet1', 108, first_col_iterate = 7)
sheet2 = melt_signals(path, 'Sheet2', 118)
sheet3 = melt_signals(path, 'Sheet3', 98)
```

```python
def upload_frames(filepath, dataframes, names = []):
    writer = pd.ExcelWriter(filepath)

    # Write each DataFrame to a separate spreadsheet
    k = 1
    for i in dataframes:
        i.to_excel(writer, sheet_name=f'Sheet{k}')
        k += 1
    writer.save()
        
```
